```python
#here totalCharges column is showing as object types but as it contains float values with some blank spaces.
#So after checking the tenure at the blank spaces, the tenure is 0.
# we are replacing the blank spaces with 0 in totalCharges column and converting its datatype to float
```

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
#we will convert the values from senior citizen column
```

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"
    
df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

```python
df
```

```python
ax = sns.countplot(x = 'Churn', data = df)

ax.bar_label(ax.containers[0])
plt.title("Count Of Customers by Churn")
plt.show()
```

```python
gb = df.groupby("Churn").agg({"Churn":"count"})
plt.pie(gb["Churn"], labels = gb.index, autopct = "%1.2f%%")
plt.title("Percentage of Churned Customers")
plt.show()
```

```python
#by looking at the above charts we can say that 26.54% of customers have churned out
#now lets explore the reason behind it
```

```python
sns.countplot("gender", data = df, hue = "Churn")
plt.title("Churn by Gender")
plt.show()
```

```python
sns.countplot("SeniorCitizen", data = df, hue = "Churn")
plt.title("Churn by SeniorCitizen")
plt.show()
```

```python
grouped_data = df.groupby(['SeniorCitizen', 'Churn']).size().reset_index(name='Count')

# Step 2: Pivot the data to have SeniorCitizen as index and Churn as columns
pivot_data = grouped_data.pivot(index='SeniorCitizen', columns='Churn', values='Count')

# Step 3: Calculate percentages for each SeniorCitizen group
pivot_data_percentage = pivot_data.apply(lambda x: 100 * x / x.sum(), axis=1)

# Step 4: Plot the stacked bar chart
ax = pivot_data_percentage.plot(kind='bar', stacked=True, color=['#1f77b4', '#ff7f0e'], figsize=(8,6))

# Step 5: Add percentage labels on each section
for p in ax.patches:
    width = p.get_width()    # Get the width of each bar segment
    height = p.get_height()  # Get the height (which is the percentage value)
    x = p.get_x() + width / 2  # Calculate x position
    y = p.get_y() + height / 2 # Calculate y position
    ax.annotate(f'{height:.1f}%', (x, y), ha='center', va='center', fontsize=10, color='white')

# Step 6: Customize the plot
plt.title('Percentage of Churn by SeniorCitizen')
plt.ylabel('Percentage')
plt.xlabel('SeniorCitizen')
plt.legend(title='Churn', loc='upper right')
plt.show()
```

```python
#from the above charts we can say that Senior citizen category has churned with high percentage
```
